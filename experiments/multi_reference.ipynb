```python
ds = ['MTG', 'ALM', 'VISp']
ds = {s: sc.read_h5ad(f'../data/{s}.h5ad') for s in ds}
```

```python
q, ref = 'MTG', 'ALM'

rcm = RefCM()
m = rcm.annotate(ds[q], q, ds[ref], ref, 'labels34', 'labels34')
m.eval('labels34')
m.display_matching_costs('labels34')
```

```python
q = 'MTG'
ref_ids = [s for s in ds if s != q]

rcm = RefCM()
m = rcm.annotate(ds[q], q, [ds[r] for r in ref_ids], ref_ids, ['labels34', 'labels34'], 'labels34')
m.eval('labels34')
m.display_matching_costs('labels34')
```
