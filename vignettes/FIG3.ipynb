```python
import scanpy as sc
import benchdb as bdb
import benchutils as bu
import benchplots as bp
import benchmodels as bm

from tqdm import tqdm 


%load_ext autoreload
%autoreload 2
```

```python
db = bdb.load_benchdb("bench.json")
bdb.mvote_all(db, to_mvote=bm.RM_METHODS)

alm = sc.read_h5ad('../data/ALM.h5ad')
bu.prep_adata(alm)
bu.prep_umap(alm)

mtg = sc.read_h5ad('../data/MTG.h5ad')
bu.prep_adata(mtg)
bu.prep_umap(mtg)
```

# FIG 3A

```python
fig = bp.umap(alm, 'labels34', title="ALM", width=850, height=750)
fig.update_layout(margin=dict(t=60), title_font_size=40)
fig.update_layout(
    title=dict(
        y=0.98,              
        yanchor="top",
        pad=dict(t=0, b=0) 
    )
)
fig.write_image("FIG3/ALM.png", scale=3)
fig.show()
```

```python
fig = bp.umap(mtg, 'labels34', title="MTG", width=850, height=750)
fig.update_layout(margin=dict(t=60), title_font_size=40)
fig.update_layout(
    title=dict(
        y=0.98,              
        yanchor="top",
        pad=dict(t=0, b=0) 
    )
)
fig.write_image("FIG3/MTG.png", scale=3)
fig.show()
```

# FIG 3B

```python
m = "MV-CellTypist"
fig = bp.umap_acc(mtg, db[m]['MTG']['ALM'], model=m, width=750, height=750)
fig.update_layout(margin=dict(t=70), title=f"{m}", title_font_size=50)
fig.update_layout(
    title=dict(
        y=0.98,              
        yanchor="top",
        pad=dict(l=0, t=0, b=0) 
    )
)
fig.show()
```

```python
for m in tqdm(bm.RCM_METHODS + bm.MV_METHODS + bm.RM_METHODS):

    a = db[m]['MTG']['ALM']
    fig = bp.umap_acc(mtg, a, model=m, width=750, height=750)
    # fig.update_layout(margin=dict(t=60), title=f"{m} | {a.acc:.2f}", title_font_size=40)
    # fig.update_layout(margin=dict(t=60), title=m, title_font_size=40)
    fig.update_layout(margin=dict(t=70), title=f"{m} | {a.acc:.2f}", title_font_size=50)
    # fig.update_layout(margin=dict(t=80), title_font_size=30)
    fig.update_layout(
        title=dict(
            y=0.98,              
            yanchor="top",
            pad=dict(t=0, b=0) 
        )
    )
    
    fig.write_image(f"FIG3/{m}.png", scale=3)
```
